```python
def create_movie_user_rating_matrix(ratings_df):
    movie_user_matrix = pd.pivot_table(ratings_df, values='Rating', index='MovieID', columns='UserID')

    return movie_user_matrix
```

```python
def users_who_rated(movieID,movie_user_matrix):
    user_ratings = movie_user_matrix.loc[movieID].dropna()
    return user_ratings.to_dict()
```

```python
def users_who_rated_intersection(userIDs_1,userIDs_2):

    np_userIDs_1 = np.array(list(userIDs_1))
    np_userIDs_2 = np.array(list(userIDs_2))

    return np.intersect1d(np_userIDs_1,np_userIDs_2)
```

```python
def similarity(new_user_ratings,user_ratings):   #assuming input as Series objects
    
    if len(new_user_ratings) == 0:     #no users in common
        return 0
    
    new_user_mean = new_user_ratings.mean()
    user_mean = user_ratings.mean()


    new_user_ratings -= new_user_mean
    user_ratings -= user_mean

    num = np.dot(new_user_ratings, user_ratings)

    new_user_ratings_square = new_user_ratings**2
    user_ratings_square = user_ratings**2

    term1 = np.sqrt(np.sum(new_user_ratings_square))
    term2 = np.sqrt(np.sum(user_ratings_square))

    if term1 == 0 and term2 == 0:
        return 1
    elif term1 == 0 or term2 == 0:
        return 0

    return num/(term1*term2)
```

```python
ageGroupMap = {}
ageGroupMap[1] = (18+1)/2
ageGroupMap[18] = (18+24)/2
ageGroupMap[25] = (25+34)/2
ageGroupMap[35] = (35+44)/2
ageGroupMap[45] = (45+49)/2
ageGroupMap[50] = (50+55)/2
ageGroupMap[56] = (56+70)/2
```

```python
ageWeights = {}
sigma = 25

ages = {1,18,25,35,45,50,56}

for user_age in range(1,100):
    if user_age not in ageWeights.keys():
        ageWeights[user_age] = {}
    for age in ages:
        ageWeights[user_age][age] = np.exp(-((user_age - int(ageGroupMap[age])) ** 2) / (2 * sigma ** 2))

print(ageWeights[21])
```

```python
def age_weighted_similarity(new_user_ratings,user_ratings,userIDs,user_df,userAge):   #assuming input as Series objects

    if len(new_user_ratings) == 0:    #if there are no common movies, the similarity of two users will be 0
        return 0

    new_user_mean = new_user_ratings.mean()
    user_mean = user_ratings.mean()

    np_new_user_ratings = new_user_ratings.values
    np_user_ratings = user_ratings.values

    new_user_ratings -= new_user_mean
    user_ratings -= user_mean

    user_weights = []
    for userID in userIDs:
        user_row = user_df[user_df['UserID'] == userID]
        user_weights.append(ageWeights[userAge][int(user_row['Age'].iloc[0])])

    num = sum(x * y * z for x, y, z in zip(user_weights, np_user_ratings, np_new_user_ratings))

    new_user_ratings_square = new_user_ratings**2
    user_ratings_square = user_ratings**2

    term1 = np.sqrt(np.dot(new_user_ratings_square,user_weights))
    term2 = np.sqrt(np.dot(user_ratings_square,user_weights))


    if term1 == 0 and term2 == 0:
        return 1
    elif term1 == 0 or term2 == 0:
        return 0

    return num/(term1*term2)
```

```python
#SImilarity computed based on the user demographic(here, gender)

merged_df = pd.merge(ratings_df, user_df, on='UserID', how='inner')

average_ratings = merged_df.groupby(['MovieID', 'Gender'])['Rating'].mean().unstack()

total_average_rating = ratings_df['Rating'].mean()

ratings_counts = merged_df.groupby(['MovieID', 'Gender']).size()

bayesian_averages = (average_ratings * ratings_counts + total_average_rating) / (ratings_counts + 1)



item_similarity = bayesian_averages.dot(bayesian_averages.T)

item_similarity = item_similarity.fillna(0)



min_value = item_similarity.min().min()
max_value = item_similarity.max().max()
normalized_item_similarity = (item_similarity - min_value) / (max_value - min_value)

np.fill_diagonal(normalized_item_similarity.values, 1)

normalized_item_similarity
```

```python
def new_rating(movieID,new_movie_ratings,movie_user_matrix,user_df,userAge):   #movieID:int, new_movie_ratings : dictionary, user_movie_matrix : dataframe

    movie_row = movie_user_matrix.loc[movieID]
    mean = movie_row.mean()

    num = 0
    den = 0

    movie_ratings = users_who_rated(movieID,movie_user_matrix)

    for i in new_movie_ratings.keys():
        
        i_ratings = users_who_rated(i,movie_user_matrix)   #returns a dictionary with userIDs as keys and their respective ratings for the given movie as values

        common_users = users_who_rated_intersection(i_ratings,movie_ratings)
        common_i_ratings = [i_ratings[user] for user in common_users]
        common_i_ratings = pd.Series(common_i_ratings)
        common_movie_ratings = [movie_ratings[user] for user in common_users]
        common_movie_ratings = pd.Series(common_movie_ratings)
       
        # sim = similarity(common_i_ratings,common_movie_ratings)
        sim = age_weighted_similarity(common_i_ratings,common_movie_ratings,common_users,user_df,userAge)
        
        new_movie_row = movie_user_matrix.loc[i]
        new_mean = new_movie_row.mean()
        num = num + sim*(new_movie_ratings[i]- new_mean)
        den = den + abs(sim)

    if den == 0:
        return mean
        
    return mean + num/den

    
```

```python
def watched_movies(userID,movie_user_matrix):
    user_ratings = {}
    for index,row in movie_user_matrix.iterrows():
        if pd.notna(row[userID]):  
            user_ratings[index] = row[userID]
    return user_ratings
```

```python
def take_n_items(dictionary, n):
    return dict(islice(dictionary.items(), n))
```

```python
def split_ratings(user_ratings):
    ratings_list = list(user_ratings.items())
    
    # Shuffle the ratings
    np.random.shuffle(ratings_list)
    # Calculate the split index for 80% training and 20% input
    split_index = int(0.8 * len(ratings_list))
    # Split the ratings
    input_ratings = dict(ratings_list[:split_index])
    test_ratings = dict(ratings_list[split_index:])
    return input_ratings, test_ratings
```

```python
def mse(ratings1,ratings2):
    mse = 0
    for i in range(len(ratings1)):
        mse = mse + (ratings1[i]-ratings2[i])**2
    mse = mse / len(ratings1)*100
    return mse
```

```python
def get_genres_for_movie(movie_df, movie_id):
    
    movie_row = movie_df[movie_df['MovieID'] == movie_id]
    
    if not movie_row.empty:
        genres_str = movie_row['Genre'].values[0]
        genres_list = genres_str.split('|')    #'Genre' column contains genres separated by '|'
        return genres_list
    else:
        return None
```
